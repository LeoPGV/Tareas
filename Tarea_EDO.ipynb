```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

def f1(x,y):
  return -2*y/x + 3/x**3 + 1

def f2(x):
  return (9*np.log(x) + x**3 + 2)/(3*x**2)

def euler(f, g, x0, y0, h, n):
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    x[0] = x0
    y[0] = y0
    m = np.zeros(n+1)
    er = np.zeros(n+1)
    ep = np.zeros(n+1)


    for i in range(1, n+1):
        y[i] = y[i-1] + h * f(x[i-1], y[i-1])
        x[i] = x[i-1] + h

    for i in range(1,n+1):
      m[i] = f2(x[i])
      er[i] = abs((f2(x[i])-y[i])/f2(x[i]))
      ep[i] = abs((f2(x[i])-y[i])/f2(x[i]))*100

    df = pd.DataFrame({
        "t": x,
        "yi": y,
        "ys": m,
        "eb":er,
        "ep":ep

    })

    plt.plot(x, y)
    plt.plot(x, m)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Método de Euler')
    plt.show()

    return df
def runge_kutta_4(f, f2, y0, t0, tf, h):


    t = np.arange(t0, tf + h, h)


    y = np.zeros(len(t))
    y[0] = y0
    m = np.zeros(len(t))
    er = np.zeros(len(t))
    ep = np.zeros(len(t))



    for i in range(1, len(t)):


        k1 = h * f(t[i-1], y[i-1])
        k2 = h * f(t[i-1] + h/2, y[i-1] + k1/2)
        k3 = h * f(t[i-1] + h/2, y[i-1] + k2/2)
        k4 = h * f(t[i], y[i-1] + k3)


        y[i] = y[i-1] + (k1 + 2*k2 + 2*k3 + k4)/6

    for i in range(0,len(t)):
      m[i] = f2(t[i])
      er[i] = abs((f2(t[i])-y[i])/f2(t[i]))
      ep[i] = abs((f2(t[i])-y[i])/f2(t[i]))*100



    df = pd.DataFrame({
        "t": t,
        "yi": y,
        "y(t)": m,
        "Er abs":er,
        "Er poc":ep

    })

    plt.plot(t, y)
    plt.plot(t, m)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Método de  Kutta')
    plt.show()

    return df
```

```python
print(euler(f1,f2, 1, 1, 0.1, 10))
print(runge_kutta_4(f1,f2, 1, 1, 2, 0.1))
```

```python
print(euler(f1,f2, 1, 1, 0.05, 20))
print(runge_kutta_4(f1,f2, 1, 1, 2, 0.05))
```

```python
print(euler(f1,f2, 1, 1, 0.025, 40))
print(runge_kutta_4(f1,f2, 1, 1, 2, 0.025))
```
