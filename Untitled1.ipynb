```python
import numpy as np
import matplotlib.pyplot as plt
import math as mt
     

def fun(start, end, ylim=2, div=50):
  X = np.linspace(start, end, div)
  Y = np.cos(X) - X * np.exp(X)
  plt.plot(X,Y)
  plt.title(f"f(x) = cos(x) - xe^x   [{start}, {end}]")
  plt.grid()
  plt.axhline(0, color="black")
  plt.axvline(0, color="black")
  plt.xlim(start, end)
  plt.ylim(-ylim, ylim)
  plt.show()
```
